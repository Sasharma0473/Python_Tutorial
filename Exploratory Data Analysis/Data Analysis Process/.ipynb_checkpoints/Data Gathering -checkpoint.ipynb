# Opening a CSV File from URL

```python
# If some csv file is on github server then we can import that file like this. Here we are working as a client.
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url,headers = headers)
data = StringIO(req.text)
country_data = pd.read_csv(data)
country_data
```

```python
import requests
from io import StringIO

url= "https://github.com/codeforamerica/ohana-api/blob/master/data/sample-csv/addresses.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers = headers)
add_data = StringIO(req.text)
new_add_data = pd.read_csv(add_data,encoding='utf-8',on_bad_lines='skip',sep='\t')
new_add_data
```

# TO CVS FILE

```python
delivery_data = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\deliveries.csv")
delivery_data
```

```python
# It will give the bastman vise total runs for all ipl matches.
temp_df = delivery_data.groupby('batsman')['batsman_runs'].sum().reset_index()
temp_df
```

```python
# this will generate the csv file. This file has index values with the data so to remove the index value we
# have to give index as false
temp_df.to_csv('batsman_runs.csv',index = False)
```

```python
# it will give the batsman runs for every batsman against each bowling team.
delivery_data.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs', aggfunc='sum').to_csv("Batsman_vs_Team")
```

# EXPORT IN EXCEL FILE

```python
df_data = delivery_data.groupby('batsman')['batsman_runs'].sum().reset_index()
df_data1 = delivery_data.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs', aggfunc='sum')
```

```python
df_data.to_excel("batsman_run_excel.xlsx")
```

```python
# we can rename the sheet with the help of this parameter. 
df_data.to_excel("batsman_run_excel.xlsx",sheet_name = 'batsman_runs')
```

```python
# here we are storing more sheet in the single excel file 

with pd.ExcelWriter('batsman_run_excel_file.xlsx') as writer:
    df_data.to_excel(writer,sheet_name='sheet_name1')
    df_data1.to_excel(writer,sheet_name='sheet_name2')
```

# EXPORT FILE IN HTML

```python
def plot_graphs(column_name):
    sns.displot(kind='hist',kde=True,data=phone_data,x=column_name,label=column_name)
    sns.catplot(kind='box',data = phone_data, x = column_name)
```

```python
phone_data.select_dtypes(include=['float64','int64']).head(2)
```

```python
num_columns = phone_data.select_dtypes(include=['float64','int64']).iloc[:,[3,4,6,9,13,14,16]].columns
```

```python
for col in num_columns:
    plot_graphs(col)
```

```python
# KNN imputor
x_df = phone_data.select_dtypes(include=["int64","float64"]).drop(columns='price')
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=5)
```

```python
x_df_values = imputer.fit_transform(x_df)
```

```python
pd.DataFrame(x_df_values, columns=x_df.columns)
```

```python
x_df1 = pd.DataFrame(x_df_values, columns=x_df.columns)
```

```python
x_df1['price']= phone_data['price']
```

```python
x_df1
```

```python
a = x_df1.corr(numeric_only=True)['price'].reset_index()
```

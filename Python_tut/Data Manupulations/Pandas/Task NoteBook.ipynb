```python
# Find batsman in below category-
# Highest score while chasing
# Chasing mean team batting in second inning
ipl_dataset = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\ipl_deliveries.csv")
ipl_innings_data = ipl_dataset[ipl_dataset["innings"]==2]
ipl_innings_data.groupby(["ID","batter"]).sum().sort_values(by="batsman_run",ascending = False)['batsman_run'].head(5)
```

```python
# Best Strike rate while chasing and have faced 100+ balls
ipl_data = ipl_innings_data[~(ipl_innings_data['extra_type']=="wides")]
ipl_data = ipl_data.groupby("batter").agg({
            "batsman_run":"sum",
            "ballnumber":"count"
})
ipl_data["strike_rate"]= ipl_data["batsman_run"]/ipl_data["ballnumber"]*100
ipl_data[ipl_data["ballnumber"]>=100].sort_values("strike_rate",ascending=False).reset_index().head()
```

```python
#Most Successful bowler against any batsman. Find that pair of bowler and batsman.
ipl_dataset["isbatterout"] = ipl_dataset['batter']==ipl_dataset['player_out']
ipl_dataset.groupby(["bowler","batter"]).agg({
    "isbatterout":"sum",
    "batsman_run":"sum"
    
}).sort_values(by=["isbatterout","batsman_run"],ascending =[False,True]).head()
```

```python
# Most successful batting pair in IPL. Batting pair who have scored most runs playing together.
def func(x):
    return '-'.join(list(np.sort(x.values)))
    
ipl_dataset["batter-pair"]=ipl_dataset[["batter","non-striker"]].apply(func,axis=1)
ipl_dataset.groupby("batter-pair")["total_run"].sum().sort_values(ascending=False).head()
```

```python
# Make a dataframe for all batting pairs played together.
# Batsman1 Batsman2 Runs Avg StrikeRate
ipl_dataset_data = ipl_dataset.groupby("batter-pair").agg({
    "total_run":"sum",
    "ballnumber":"count",
    "isWicketDelivery":"sum"
}).reset_index()
ipl_dataset_data["Batsman 1"]=ipl_dataset_data["batter-pair"].apply(lambda x:x.split("-")[0])
ipl_dataset_data["Batsman 2"]=ipl_dataset_data["batter-pair"].apply(lambda x:x.split("-")[1])
ipl_dataset_data.rename(columns = {"total_run": "Runs"}, inplace=True)
ipl_dataset_data["StrikeRate"] = ipl_dataset_data["Runs"] / ipl_dataset_data["ballnumber"] * 100
ipl_dataset_data["Avg"] = ipl_dataset_data["Runs"] / ipl_dataset_data["isWicketDelivery"]  # runs per wicket
ipl_dataset_data.sort_values("Runs", ascending = False, inplace=  True)
ipl_dataset_data[["Batsman 1", "Batsman 2", "Runs", "Avg", "StrikeRate"]]
```

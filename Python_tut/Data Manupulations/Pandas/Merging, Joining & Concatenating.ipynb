```python
# 1. find total revenue generated
regs.merge(course,how='inner',on='course_id')['price'].sum()
```

```python
# Have a index as Q-1 Q-2 Q-3 for respective quater files in the dataframe
total_quarter = pd.concat([quart1,quart2,quart3], keys = ['q1','q2','q3']).reset_index()
total_quarter.shape
total_quarter
```

```python
# Your are given a file items.csv which has item_id and item_name. Find out most sold items in each quarter.
sold_item = total_quarter.merge(item_data, how = 'left', on ='item_id').reset_index()
# print(sold_item)
sold_item.rename(columns={"level_0":"quarter"},inplace= True)
# print(sold_item)
sold_item.groupby(["quarter","item_name"])['quantity'].sum().reset_index().sort_values('quantity', ascending = False).drop_duplicates(subset=['quarter'], keep = 'first')                              
```

```python
# Find out items which has made most revenue in each quarter.
# sold_item = sold_item.drop('total_price',axis=1)
sold_item['item_price'] = sold_item['item_price'].apply(lambda x: float(x[1:]))
sold_item['total_price'] = sold_item['quantity'] * sold_item['item_price']
sold_item.groupby(['quarter','item_id'])['total_price'].sum().reset_index().sort_values('total_price',ascending = False).drop_duplicates(subset = ['quarter'], keep = 'first')                                
```

```python
# # Find out avg order price of each quarter.
sold_item.groupby(['quarter','order_id'])['total_price'].sum().reset_index().groupby('quarter')['total_price'].mean().reset_index()
```

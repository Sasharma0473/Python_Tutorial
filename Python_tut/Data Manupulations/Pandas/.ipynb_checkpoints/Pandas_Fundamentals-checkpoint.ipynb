```python
delhivery = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\deliveries.csv")
delhivery.head(2)
```

```python
delhivery.info()
```

```python
# to find out top 5 batsman in terms of run score in 10 season.
runs = delhivery.groupby('batsman')
runs.get_group('V Kohli')
```

```python
runs['batsman_runs'].sum().sort_values(ascending = False).head(5)
```

```python
# find out the batsman who made more 6's in 10 years.
mask3 = delhivery['batsman_runs']==6
new_delhivery= delhivery[mask3]
```

```python
new_delhivery.shape[0]
```

```python
new_delhivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(5)
# new_delhivery.groupby('batsman')['batsman_runs'].size().sort_values(ascending=False).head(5) we can use both
```

```python
# find out the highest runs made by virat kohli against which team.
mask4 = delhivery['batsman'] == 'V kohli'
new_delhivery1 = delhivery[mask4]
new_delhivery1.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(5)
```

```python
def run_scored(batsman_name):
    mask4 = delhivery['batsman'] == batsman_name
    new_delhivery1 = delhivery[mask4]
    return new_delhivery1.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index
```

```python
run_scored('V kohli')
```

```python
# isin()
# find the most destructive death over batsman in the history of IPL. we decide this on the basis of strike rate
# strike rate = (No of runs/No of balls)*100
# min batsman 200 balls in over 16-20
death_over = delhivery[delhivery['over']>15]   # Same as masking
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()  # Balls played by batsman.
x= all_batsman > 200
batsman_list = all_batsman[x].index.tolist()
```

```python
# Runs Scored by all these 43 batsman 
# Balls played by these 43 batsman
final = delhivery[delhivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
sr = (runs/balls)*100
sr
```

# Handling Missing Values

```python
# We can use two main functions for handling the missing values.
# dropna and fillna - in these two functions it will basically used for dropping and filling the missing values in the data.
train = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\train.csv")
```

```python
train.shape
```

```python
train
```

```python
train.dropna()
```

```python
train.dropna().shape
```

```python
train.dropna(how='all').shape   
# how parameter = all means remove all rows where ALL the columns values is missing.
# so here in 891 row we can conclude that no row have all the missing values.
# axis parameter = if we put axis = 0 means for rows and axis =1 means for columns.
# subset paramerter:- It will basically remove the missing value from one or more column.

```

```python
train.dropna(subset=['Cabin','Embarked']).shape
```

```python
# Filling the missing values through fillna()
# fillna(method='bfill or ffill') with the help of bfill and ffill 
# we can use it will fill the value from the backward and forward column
train.fillna(0)
```

```python
# In every column we can use 0 value as some column could contain value in string.
train['Cabin'].fillna("Not Specified")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
```

```python
#Create a scatter plot of 'gdpPercap' against 'lifeExp' for the year 2007, 
#with the size of the markers determined by 'pop' and the color determined by 'continent'.
gap_data = px.data.gapminder()
gap_data.head(2)
```

```python
gap_2007_data = gap_data[gap_data['year']==2007]
sns.scatterplot(data=gap_2007_data,x='gdpPercap',y='lifeExp',hue='continent',size='pop')
```

```python
# Using flights dataset of seaborn.

# Using the "flights" dataset that comes with seaborn, 
# create a heatmap that shows the average number of passengers per month for each year.

# Using the seaborn's flight dataset, create a clustermap to visualize 
# the relationship between the number of passengers, months, and year.

flight_data = sns.load_dataset('flights')
flight_data.head(2)
```

```python
f_data = flight_data.pivot_table(index='year',columns='month',values='passengers',aggfunc='mean')
sns.heatmap(f_data)
```

```python
insurance_data = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\insurance_data - insurance_data.csv")
insurance_data.head(2)
```

```python
#Draw a scatter plot based on the below conditions:
#x-axis should be "age" and y-axis should be "bmi".
#For hue, size and style parameters use "diabetic", "gender" and "smoker" column respectively.
#Add title to your chart.
#Age should be less than 70 percentiles.
#BMI should be greater than the average value of the filtered age dataset.
```

```python
insur_data = insurance_data[insurance_data['age']< insurance_data['age'].quantile(0.70)]
temp_df = insur_data[insur_data['bmi']> insur_data['bmi'].mean()]
sns.scatterplot(data= temp_df, x='age',y='bmi',hue='diabetic',style='smoker',size='claim')
```

```python
#Draw a line plot by using the below informations
#bloodpressure vs children
#Blood-pressure values should be between 90 and 100. The upper and lower limit are included.
#Show the details of "smoker".
temp_df = insurance_data[(insurance_data['bloodpressure']>=90) & (insurance_data['bloodpressure']<=100)]
sns.lineplot(data=temp_df,x='bloodpressure',y='children',hue='smoker',err_style=None)
```

```python
#Draw a histogram using displot
#based on "age" column.
#Show details of "smoker" (hue).
#Create 2 separate charts for the above 2 conditions based on "gender" side-by-side.

sns.displot(kind='hist',data=insurance_data,x='age',hue='smoker',col='gender')
```

```python
#Draw a kde plot between "age" and "bloodpressure".
sns.kdeplot(data=insurance_data,x='bloodpressure',y='age')
```

```python
#Draw a clustermap between between "age", "bmi" and "bloodpressure".
sns.clustermap(insurance_data[['age','bmi','bloodpressure']].dropna())
```

```python
taxi_data = sns.load_dataset('taxis')
taxi_data.head(2)
```

```python
sns.catplot(data=taxi_data,x='payment',y='total',kind='point')
```

```python
taxi_data['time_ride']= pd.to_datetime(taxi_data['dropoff'])-pd.to_datetime(taxi_data['pickup'])
taxi_data
```

```python
# to convert it to seconds 
taxi_data['time_ride']= round(taxi_data['time_ride'].dt.total_seconds()/60,2)
```

```python
sns.lmplot(data=taxi_data,x='time_ride',y='total')
```

```python
sns.lmplot(data=taxi_data,x='time_ride',y='total',hue='color')
```

```python
sns.lmplot(data=taxi_data,x='time_ride',y='total',hue='payment')
```

```python
 # Draw a reg plot between "age" and "bmi" columns.

sns.regplot(data=insurance_data,x='age',y='bmi')
```

```python
# Draw a pair plot of the insurance data. Use "gender" as hue parameter.
sns.pairplot(data=insurance_data.drop(columns=['index','PatientID']), hue='gender')
```

```python
#Draw a pair grid of the insurance data and use "diabetic" column as a hue parameter. 
#Also, make the diagonal plots as box-plot, upper parts as scatter plot and the lower parts as kde plot.
g = sns.PairGrid(data=insurance_data.drop(columns=['index','PatientID']),hue='diabetic')
g.map_diag(sns.boxplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

```python
# Draw a joint plot as scatter between "bloodpressure" and "bmi". Use "smoker" as hue parameter.
sns.jointplot(data = insurance_data, x='bloodpressure',y='bmi',kind='scatter',hue='smoker')
```

```python
# Draw a joint grid of which x-axis represents "age" and y-axis represents "claim". 
# Draw a kdeplot at the center and violinplot for individual variables
g = sns.JointGrid(data=insurance_data,x='age',y='claim')
g.plot(sns.kdeplot,sns.violinplot)
```

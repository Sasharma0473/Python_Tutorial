# 2D LINE PLOT

- It is build when we have to perform the analysis on 2 columns.Mostly used when we have numeric-numeric or numeric-categorical data use case - time series data. It is used when we have to do bivariate analysis.

- univariate vs bivariate vs multivariate:

- when we have to perform the analysis on single column or single data we called it univariate, when we have to 
perform analysis on 2 column or 2 data then we called it as bivariate, when we have to perform analysis on 3 columns 
is called multivariate.

```python
# Color(hex) and line(width and style) and marker(size)
plt.plot(batsman_dataset["index"],batsman_dataset["V Kohli"],color="blue")    # use hexdecimalcode to get particular color.
plt.plot(batsman_dataset["index"],batsman_dataset["RG Sharma"],color="yellow")
plt.title("Rohit Sharma VS Virat Kohli Comparison")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

```python
# line(width and style)
# linestyle = solid, dashed, dotted,dashdot.
# linewidth: Its basically used for width of line.
plt.plot(batsman_dataset["index"],batsman_dataset["V Kohli"],color="red",linestyle="solid",linewidth=3.5)
plt.plot(batsman_dataset["index"],batsman_dataset["RG Sharma"],color="blue",linestyle="dashed",linewidth=1.5)
plt.title("Rohit Sharma VS Virat kohli Comparsion")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

```python
# marker(size) 
# if we want to see the points in the graph which is corresponding to y axis.(D,+,.,>,<,o,)
# markersize parameter used for controlling the size of marker.

plt.plot(batsman_dataset["index"],batsman_dataset["V Kohli"],color="red",linestyle="solid",linewidth=3.5,marker='D',markersize=10)
plt.plot(batsman_dataset["index"],batsman_dataset["RG Sharma"],color="blue",linestyle="dashed",linewidth=1.5,marker='+',markersize=15)
plt.title("Rohit Sharma VS Virat kohli Comparsion")
plt.xlabel("Season")
plt.ylabel("Run Scored")
```

```python
# Parameter - color,linewidth,linestyle,marker,markersize,label
# legends: used for labelling the line on the graph.in the label parameter we give legend and we have to cal the legends
# by using func
# legend parameter take loc by default its best but we can change and tell legend where to appear in graph.
# loc = best,upper right 
plt.plot(batsman_dataset["index"],batsman_dataset["V Kohli"],color="yellow",linestyle="solid",linewidth=3.5,marker='D',markersize=10,label='Virat')
plt.plot(batsman_dataset["index"],batsman_dataset["RG Sharma"],color="blue",linestyle="dashed",linewidth=1.5,marker='+',markersize=15,label='Rohit')
plt.title("Rohit VS Virat Comparision")
plt.xlabel("Seasons")
plt.ylabel("Run Scored")
plt.legend(loc='best')        #NEW FUNCTION
```

```python
# limiting axis parameter: It is basically used to trim the x and y axis data in the graph. 
# We use ylim and xlim for trimming x and y axis.
# here we can see that becz of 2021 data other values is like showing a line does not give me actual data view
# so here we use xlim and ylim to limit the data for particular region in the graph.
# Mostly we use trimming of graph when we got the outlier value.

price = [45000,25000,15000,55000,60000,20000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,100000) # it will trim data on y-axis
# plt.xlim(2017,2019) # It will trim data on x- axis
```

```python
# plt.grid: to get the more clarity about the data. it will print the grid in the backgorund of graph.
# plt.show: use to print the graph when we use different console.
plt.plot(batsman_dataset["index"],batsman_dataset["V Kohli"],color="yellow",linestyle="solid",linewidth=3.5,marker='D',markersize=10,label='Virat')
plt.plot(batsman_dataset["index"],batsman_dataset["RG Sharma"],color="blue",linestyle="dashed",linewidth=1.5,marker='+',markersize=15,label='Rohit')
plt.title("Rohit VS Virat Comparision")
plt.xlabel("Seasons")
plt.ylabel("Run Scored")
plt.legend(loc='best')  # NEW FUNCTION
plt.grid()
```

# SCATTER PLOT
Its used for bivariate analysis(study about two columns).It always works on numerical-numerical data. Scatter plot is slower than the 2D plot. 

usecase: to find correlation between 2 quantites.

Difference between 2D plot and scatter plot:
Scatter plot is all about plotting the points on the graph, in this we just plot the points and do'nt join them. 2D plot is a special case of scattter plot where we 
join the lines between the points. 

```python
# Scatter plot simple func
x = np.linspace(-10,10,50)  # linspace(take values from -10 to 10, total values between the range)
y = 10*x+3 + np.random.randint(0,300,50)
# print(x)
# print(y)
plt.scatter(x,y)
```

```python
# simple bar chart 
children = [10,20,40,10,30]     # numeric value.
colors = ["red","blue","green","yellow","pink"]  # category.
plt.bar(colors,children,color="black")
plt.title("Children like different color")
plt.xlabel("colors")
plt.ylabel("No of children")
```

# SUBPLOTS
We can plot multiple graphs together. If we want to compare values then we have to plot multiple graphs.

```python
batter_data
plt.scatter(batter_data['avg'],batter_data['strike_rate'],color = 'yellow')
```

```python
fig , ax = plt.subplots(figsize=(15,7))
ax.scatter(batter_data['avg'],batter_data['strike_rate'])
ax.set_title("Avg VS Strike_rate")
ax.set_xlabel("Average")
ax.set_ylabel("Strike Rate")
fig.show()
```

```python
fig , ax = plt.subplots(figsize= (15,7), nrows=2, ncols=1,sharex=True) # plot multiple graphs.
# print(fig)
# ax
ax[0].scatter(batter_data['avg'],batter_data['strike_rate'],color='yellow') 
ax[0].set_title("Avg VS Strike Rate")
ax[0].set_ylabel("Strike Rate")
ax[1].scatter(batter_data['avg'],batter_data['runs'])
ax[1].set_title("Avg VS Runs")
ax[1].set_ylabel("Runs")
```

```python
# For sharing x axis and y axis we use sharex and sharey
fig , ax = plt.subplots(figsize= (10,10), nrows=2, ncols=2,sharex=False) # plot multiple graphs.
# print(fig)
# ax
ax[0,0].scatter(batter_data['avg'],batter_data['strike_rate'],color='yellow') 
ax[0,1].scatter(batter_data['avg'],batter_data['runs'],color = 'red')
ax[1,0].hist(batter_data['avg'])
ax[1,1].hist(batter_data['runs'])
```

```python
# add_subplots(no of rows,no of columns, the graph number)
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batter_data['avg'],batter_data['strike_rate'],color='yellow') 
ax2 = fig.add_subplot(2,2,2)
ax2.hist(batter_data['avg'])
ax3 = fig.add_subplot(2,2,3)
ax3.scatter(batter_data['avg'],batter_data['runs'])
ax4 = fig.add_subplot(2,2,4)
ax4.hist(batter_data['runs'])
```

# 3D SURFACE PLOT

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2
z.shape
fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)+np.cos(yy)
fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)+np.log(yy)
fig = plt.figure(figsize = (12,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

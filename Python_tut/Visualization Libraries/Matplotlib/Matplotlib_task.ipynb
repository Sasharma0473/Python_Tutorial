```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('default')
```

```python
population_data = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\List of most-polluted cities by particulate matter concentration - List of most-polluted cities by particulate matter concentration.csv")
population_data.head(2)
```

```python
#Add a label to every axis and add a proper title for the charts. 
#Also add proper labels if there are multiple representations. 
#Then, you can customize it as your wish.
```

```python
# Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China.
china_series= population_data[population_data['Country']=='China'].groupby('Year')['PM2.5'].sum()
Iran_series = population_data[population_data['Country']=='Iran'].groupby('Year')['PM2.5'].sum()
Iran_series
```

```python
plt.plot(Iran_series.index,Iran_series.values,label='Iran')
plt.plot(china_series.index,china_series.values,label='china')
plt.title("PM2.5 OVER YEARS")
plt.xlabel("Year")
plt.ylabel("PM2.5 SUM")
plt.grid()
plt.legend()
```

```python
# Draw a histogram of the column "PM10" of which the y-axis represents the probability 
# (see the documentation how to draw the probability).
# density is a parameter used to show the probability on y axis
plt.hist(population_data['PM10'],density=True,bins=50,alpha =1)   
plt.title("Histogram of Column PM10")
plt.xlabel("bins")
plt.ylabel("probability")
plt.grid()
```

```python
# Draw a scatter plot where x-axis represents "PM2.5" and y-axis represents "PM10" for two countries Poland and Chile.
population_data.head(2)

poland_data = population_data[population_data['Country']=='Poland']
Chile_data = population_data[population_data['Country']=='Chile']
plt.scatter(poland_data["PM2.5"],poland_data["PM10"],marker='D',label="Poland")
plt.scatter(Chile_data["PM2.5"],Chile_data["PM10"],marker='+',label="Chile")
plt.title("Poland and Chile PM2.5 and PM10 Comparision")
plt.xlabel("PM2.5")
plt.ylabel("PM10")
plt.legend()
plt.grid()
```

```python
# Draw a pie chart of top 5 most frequent countries.
top_country = population_data["Country"].value_counts().head(5)
plt.pie(top_country,labels=top_country.index,autopct='%0.1f%%',explode=[0,0,0.5,0,0])
plt.title('TOP COUNTRY DATA')
plt.show()
```

```python
top_country = population_data["Country"].value_counts().head(5)
plt.bar(top_country.index,top_country.values,color ='black')
plt.xlabel("Country")
plt.ylabel("Count of Country")
plt.title("TOP 5 COUNTRY DATA")
plt.show()
```

# TASK 2 MATPLOTLIB

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder # label encoder is used to convert categorical data into numeric data.
```

```python
power_gen_dataset = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\PowerGeneration - PowerGeneration.csv")
power_gen_dataset.head(2)
```

```python
# Problem-1: Make a subplots which have 2 plots.

#For the first chart, draw a scatter plot "Monitored Cap.(MW)" vs "Total Cap. Under Maintenance (MW)" 
#of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. 
#Change the colors according to the names of the Power Stations.

#For the second chart, draw a scatter plot "Monitored Cap. (MW)" vs "Actual(MU)" of 
#the top 5 most frequent power stations. 
#Also draw the lines which indicates the average values of these two columns. 
#Change the colors according to the names of the Power Stations.
```

```python
most_freq = power_gen_dataset['Power Station'].value_counts().head().keys().tolist()
most_freq
```

```python
top_five_freq_power = power_gen_dataset[power_gen_dataset['Power Station'].isin(most_freq)]
top_five_freq_power.sample(5)
```

```python
# label endoder is used to convert the category into some number(0,1,2,3...)
encoder = LabelEncoder()
top_five_freq_power['Power Station'] = encoder.fit_transform(top_five_freq_power['Power Station'])
```

```python
top_five_freq_power.head(2)
```

```python
fig , ax = plt.subplots(1,2,figsize=(12,5))
ax[0].scatter(top_five_freq_power['Monitored Cap.(MW)'],top_five_freq_power['Total Cap. Under Maintenace (MW)'],c =top_five_freq_power['Power Station'])
ax[0].axvline(top_five_freq_power['Monitored Cap.(MW)'].mean(),c='red')
ax[0].axhline(top_five_freq_power['Total Cap. Under Maintenace (MW)'].mean(),c='red')
ax[0].set_xlabel("Montiored Cap.")
ax[0].set_ylabel("Total Cap")
ax[0].set_title("Monitored VS Total Cap")

ax[1].scatter(top_five_freq_power['Monitored Cap.(MW)'],top_five_freq_power['Actual(MU)'],c =top_five_freq_power['Power Station'])
ax[1].axvline(top_five_freq_power['Monitored Cap.(MW)'].mean(),c='red')
ax[1].axhline(top_five_freq_power['Actual(MU)'].mean(),c='red')
ax[1].set_xlabel("Montiored Cap.")
ax[1].set_ylabel("Actual(MU)")
ax[1].set_title("Monitored VS Actual(MU)")
```

```python
# Draw a 3D Scatter plot between "Monitored Cap.(MW)", "Total Cap. Under Maintenace (MW)" and "Forced Maintanence(MW)"
fig = plt.figure()
ax = plt.subplot(projection ='3d')
ax.scatter3D(top_five_freq_power['Monitored Cap.(MW)'],top_five_freq_power['Total Cap. Under Maintenace (MW)'],top_five_freq_power['Forced Maintanence(MW)'],c =top_five_freq_power['Power Station'] )
ax.set_xlabel('Monitored Cap.(MW)')
ax.set_ylabel('Total Cap. Under Maintenace (MW)')
ax.set_zlabel('Forced Maintanence(MW)')
ax.set_title("3D plot of Scatter Plot")
```

```python
# Make a 3D Surface plot of this below mathematical equation.
# z=|x||y|
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = abs(xx)-abs(yy)
```

```python
fig = plt.figure(figsize=(5,5))
ax = plt.subplot(projection = '3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
# Draw the 3D Contour plot of this below equation:
# z=|x||y|
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = abs(xx)-abs(yy)
fig = plt.figure(figsize=(5,5))
ax = plt.subplot()
ax.contour(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
# Draw a second type of Countour plot of the below equation:
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = abs(xx)-abs(yy)

fig = plt.figure(figsize= (10,10))
ax = plt.subplot()
ax.contourf(xx,yy,z,cmap='viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.show()
```

```python
# Line plot of closing value of top 5 Stocks in Year 2020.
# Take top 5 stocks based on total turnover in Year 2020
```

```python
nifty_data = pd.read_csv(r"C:\Users\sasha\Desktop\Python\Statistics\DataSet\nifty-50 - nifty-50.csv")
nifty_data.head(2)
```

```python
nifty_data['Date'] = pd.to_datetime(nifty_data['Date'])
nifty_data.set_index('Date',inplace=True)
```

```python
nifty_data.info()
```

```python
nifty_2020_data = nifty_data[nifty_data.index.year == 2020]
```

```python
top_5_data = nifty_2020_data.groupby('Symbol')['Turnover'].sum().sort_values(ascending=False).head().index
```

```python
df_top_company = nifty_2020_data[nifty_2020_data['Symbol'].isin(top_5_data)]
for name,group in df_top_company.groupby('Symbol'):
    plt.plot(group.index,group['Close'],label=name)
plt.legend()
plt.show()
```

```python
# Scatter plot Close price vs Volume for TOP-5 Stocks in year 2021
nifty_2021_data = nifty_data[nifty_data.index.year == 2021]
top_5_2021_data = nifty_2021_data.groupby('Symbol')['Turnover'].sum().sort_values(ascending=False).head().index
top_company_2021 = nifty_2021_data[nifty_2021_data['Symbol'].isin(top_5_2021_data)]
```

```python
top_company_2021['Symbol'] = top_company_2021['Symbol'].astype('category')
top_company_2021.plot(kind='scatter',x='Close',y = 'Volume', c= 'Symbol',cmap='viridis')
```

```python
# Create a 3-D Scatter Plot using time,x,y on below synthetic data. and give color gradiant on z
# Create a 3D dataset
time = np.linspace(0, 10, 100)
x = np.sin(time)
y = np.cos(time)
z = time
# Create a DataFrame from the dataset
data = pd.DataFrame({'time': time, 'x': x, 'y': y, 'z': z})
data
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter(data['time'],data['x'],data['y'],c = data['z'])
plt.title("Vibration Characterstic of car body")
plt.show()
```

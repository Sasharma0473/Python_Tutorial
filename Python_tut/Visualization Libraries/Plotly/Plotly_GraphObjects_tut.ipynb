```python
ipl_data = delivery.merge(match_data,left_on='match_id',right_on='id')
ipl_data.head(2)
```

# SCATTER PLOT

```python
#scatter plot are drawn between 2 continous variable.
# we are going to draw a scatter plot between batsman Avg(X) and batsman StrikeRate(Y) of the top 50 batsman.
top50= ipl_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(50).index.tolist()
new_ipl_data = ipl_data[ipl_data['batsman'].isin(top50)]
```

```python
# calculating the Strike Rate.
# SR = ([No of Runs Scored/No of balls played])*100
runs = new_ipl_data.groupby('batsman')['batsman_runs'].sum()
balls = new_ipl_data.groupby('batsman')['batsman_runs'].count()
sr = (runs/balls)*100
sr = sr.reset_index()
sr.head(2)
```

```python
# Calculating Average 
# Avg = Total no of runs/no of out
out = ipl_data[ipl_data['player_dismissed'].isin(top50)]
nouts = out['player_dismissed'].value_counts()
avg = runs/nouts
avg = avg.reset_index()
avg.rename(columns= {'index':'batsman',0:"avg"},inplace = True)
avg = avg.merge(sr,on='batsman')
avg
```

```python
# Plot Scatter Plot.
trace1 = go.Scatter(x = avg['avg'],y=avg['batsman_runs'],mode='markers',text = avg['batsman'],marker={"color":"green","size":16})
data = [trace1]
layout = go.Layout(title='Batsman Avg VS Strike Rate',
                   xaxis = {'title':'batsman Avg'},
                   yaxis = {"title":"Strike Rate"})
fig = go.Figure(data,layout)
pyo.plot(fig,filename='myplot.html')
```

# LINE CHARTS
Its the extension of scatter plot. Usually used to show the time series data.

```python
# year by year batsman performance.
single = ipl_data[ipl_data['batsman']=='V Kohli']
performance = single.groupby('season')['batsman_runs'].sum().reset_index()
performance
```

```python
single1 = ipl_data[ipl_data['batsman']=='MS Dhoni']
performance1 = single1.groupby('season')['batsman_runs'].sum().reset_index()
performance1
```

```python
# We can use mode as lines,marker, lines+markers
trace1 = go.Scatter(x = performance['season'],y = performance['batsman_runs'],
                    mode='lines+markers',marker = {'color':'green'},name='Virat')
trace2 = go.Scatter(x=performance1['season'],y=performance1['batsman_runs'],
                   mode = 'lines+markers',marker = {'color' : 'red'},name = 'MS Dhoni')

data =[trace1,trace2]
layout = go.Layout(title = "Year by Year Performance", xaxis = {"title":"Season"},yaxis = {"title":"Total Runs"})
fig = go.Figure(data,layout)
pyo.plot(fig)
```

```python
# Multiple Line charts 
# *name is basically a positional argument which take upto n number of different argument.
def batsman_comp(*name):
    data =[]
    for i in name:
        single = ipl_data[ipl_data['batsman']== i]
        performance = single.groupby('season')['batsman_runs'].sum().reset_index()
        trace = go.Scatter(x = performance['season'],y = performance['batsman_runs'],
                    mode='lines+markers',name=i)
        data.append(trace)
    layout = go.Layout(title = "Year by Year Performance", xaxis = {"title":"Season"},yaxis = {"title":"Total Runs"})
    fig = go.Figure(data=data,layout=layout)
    pyo.plot(fig,filename='Year-by-Year.html')
batsman_comp('MS Dhoni','V Kohli','RG Sharma','SK Raina')
```

# BAR PLOT 
It basically used to plot one categorical data value and other numerical data value.

```python
top10 = ipl_data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending = False).head(10).index.tolist()
top10df = ipl_data[ipl_data['batsman'].isin(top10)]
top10df.head(2)
```

```python
top10_score= top10df.groupby('batsman')['batsman_runs'].sum().reset_index()
top10_score
```

```python
# plot the bar graph
trace = go.Bar(x= top10_score['batsman'],y=top10_score['batsman_runs'])
data=[trace]
layout = go.Layout(title = "Top 10 batsmans", xaxis ={"title":"batsman"},yaxis={"title":"Batsman_Runs"})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# 3 Types of BAR GRAPHS
1). Nested Bar Graph
2). Stacked Bar Graph
3). OverLayed Bar Graph.

```python
# There are 3 types of Bar garph 
iw = top10df.groupby(['batsman','inning'])['batsman_runs'].sum().reset_index()
# print(iw)
one_inning = iw[iw['inning']==1]
two_inning = iw[iw['inning']==2]
one_inning.rename(columns = {"batsman_runs":"1st Innings"},inplace=True)
two_inning.rename(columns= {"batsman_runs":"2nd Innings"},inplace=True)
final_data = one_inning.merge(two_inning,on='batsman')[['batsman','1st Innings','2nd Innings']]
final_data
```

```python
# In layout we have barmode parameter which is used to represent what kind of bar graph needed.
# We can use barmode as overlay,stack, and if we do'nt give barcode then it will generate Nested bar graph.
trace1 = go.Bar(x=final_data['batsman'],y=final_data['1st Innings'],name='1st Innings',marker={"color":"green"})
trace2 = go.Bar(x=final_data['batsman'],y= final_data['2nd Innings'],name = '2nd Innings', marker={"color" : "red"})
data = [trace1,trace2]
layout = go.Layout(title="Batsman Runs per Inning",xaxis={"title":"batsman"},yaxis={"title":"Innings"})
fig = go.Figure(data = data,layout = layout)
pyo.plot(fig)
```

# 4). BUBBLE PLOT
again extension of scatter plot, with some additional information.

```python
new_ipl = new_ipl_data[new_ipl_data['batsman_runs']==6]
six = new_ipl.groupby('batsman')['batsman_runs'].count().sort_values(ascending = False).reset_index()
x = avg.merge(six,on ='batsman')
x.rename(columns={"batsman_runs_y":"No of Six"})

trace = go.Scatter(x=x['avg'],y=x['batsman_runs_x'], mode = 'markers', marker={'size':x['batsman_runs_y']})
data = [trace]
layout = go.Layout(title = "Bubble Chart",
                  xaxis = {"title":"Average"},
                  yaxis = {"title":"Strike Rate"}
                  )
fig = go.Figure(data = data , layout = layout)
pyo.plot(fig)
```

# 5). BOX PLOT 
It will Display the 5 number Summary of the set of data.

```python
match_agg = delivery.groupby('match_id')['total_runs'].sum().reset_index()
season_wise = match_agg.merge(match_data,left_on = 'match_id',right_on='id')[['match_id','total_runs','season']]
season_wise
```

```python
trace = go.Box(x=season_wise['total_runs'],name="All Seasons",marker={"color":"green"})
data = [trace]
layout = go.Layout(title="Total Score Analysis",
                  xaxis = {"title":"Total Score"})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
trace1 = go.Box(x=season_wise[season_wise['season']==2017]['total_runs'],name="2017",marker={"color":"green"})
trace2 = go.Box(x=season_wise[season_wise['season']==2008]['total_runs'],name = "2008", marker={"color":"red"})
data = [trace1,trace2]
layout = go.Layout(title="Total Score Analysis",
                  xaxis = {"title":"Total Score"})
fig = go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

# DISTPLOTS

```python
# plot distplot 
import plotly.figure_factory as ff
hist_data = [avg['avg'],avg['batsman_runs']]
group_labels=['Average',"Strike Rate"]
fig = ff.create_distplot(hist_data,group_labels,bin_size=[10,20])
pyo.plot(fig)
```

# HISTOGRAMS

```python
x = delivery.groupby('batsman')['batsman_runs'].count()>150
x = x[x].index.tolist()
new = delivery[delivery['batsman'].isin(x)]
runs = new.groupby('batsman')['batsman_runs'].sum()
balls = new.groupby('batsman')['batsman_runs'].count()
sr = (runs/balls)*100
sr = sr.reset_index()
sr
```

```python
trace = go.Histogram(x=sr['batsman_runs'],xbins={"size":2,"start":50,"end":100})
data = [trace]
layout = go.Layout(title='Strike Rate Analysis',xaxis = {"title":"Strike Rate"},yaxis={"title":"frequency Count"})
fig = go.Figure(data=data,layout = layout)
pyo.plot(fig)
```

# HEATMAPS

```python
six = delivery[delivery['batsman_runs']==6]
six = six.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()
six
```

```python
trace = go.Heatmap(x=six['batting_team'],y=six['over'],z=six['batsman_runs'])
data = [trace]
layout = go.Layout(title= "six heatmap")
fig= go.Figure(data=data,layout=layout)
pyo.plot(fig)
```

```python
# Side by Side heatmap
dots = delivery[delivery['batsman_runs']==0]
dots = dots.groupby(['batting_team','over'])['batsman_runs'].count().reset_index()

from plotly import tools

trace1 = go.Heatmap(x = dots['batting_team'],y=dots['over'],z=dots['batsman_runs'].values.tolist())
trace2 = go.Heatmap(x=six['batting_team'],y=six['over'],z=six['batsman_runs'].values.tolist())

fig = tools.make_subplots(rows=1,cols=2,subplot_titles=["6's","0's"],shared_yaxes=True)
fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)

pyo.plot(fig)
```

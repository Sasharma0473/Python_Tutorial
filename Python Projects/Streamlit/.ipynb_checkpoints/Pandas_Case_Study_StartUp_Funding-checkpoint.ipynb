```python
startup_data2['Investors Name'] = startup_data2['Investors Name'].fillna('Undisclosed')
```

```python
startup_data2.info()
```

```python
investor_data = sorted(startup_data2['Investors Name'].unique().tolist())
```

```python
startup_data2.sample(2)
```

```python
# In remark we have only 419 values filled so here we will drop that columns
startup_data2.drop(columns=['Remarks'],inplace= True)
```

```python
# here we want to change the S.No columns as a index.
startup_data2.set_index('Sr No',inplace= True)
```

```python
# Renaming the columns 
startup_data2.rename(columns={
    "Date dd/mm/yyyy":'date',
    "Startup Name": "startup",
    "Industry Vertical": "vertical",
    "SubVertical": "sub vertical",
    "City  Location": "city",
    "Investors Name":"investor",
    "InvestmentnType":"round",
    "Amount in USD": "amount"
}, inplace= True)
```

```python
startup_data2.info()
```

```python
startup_data2['amount'] = startup_data2['amount'].fillna(0)
startup_data2
```

```python
# we can'nt change this in int because we have values like 2,00,000 so we have to remove comma.
startup_data2['amount']= startup_data2['amount'].str.replace(',','')
# here in data we have some undisclosed value also so will have to replace it with 0
startup_data2['amount']= startup_data2['amount'].str.replace('undisclosed','0')
# here in data we have some unknown value also so will have to replace it with 0
startup_data2['amount']= startup_data2['amount'].str.replace('unknown','0')
startup_data2['amount']= startup_data2['amount'].str.replace('Undisclosed','0')
```

```python
startup_data2.info()
```

```python
startup_data2['amount']= pd.to_numeric(startup_data2['amount'], errors='coerce')
startup_data2.info()
```

```python
startup_data2
```

```python
def to_inr(dollar):
    inr = dollar*82.2
    return inr/10000000
```

```python
to_inr(20000000)
```

```python
startup_data2['amount'] = startup_data2['amount'].apply(to_inr)
```

```python
startup_data2.head(2)
```

```python
startup_data2['date']= pd.to_datetime(startup_data2['date'],errors='coerce')
```

```python
startup_data2.info()
```

```python
startup_data2= startup_data2.dropna(subset=['date','startup','vertical','city','investor','round','amount'])
```

```python
startup_data2.info()
```

```python
startup_data2.to_csv('startup_cleaned_data.csv')
```
